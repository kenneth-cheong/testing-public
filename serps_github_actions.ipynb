```python
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)
```

```python
with open('output.txt') as f:
    lines = f.readlines()
lines
```

```python
len(lines)
```

```python
#list of indexes of urls
new = [i for i, s in enumerate(lines) if 'http' in s]

#list of indexes of title
indexes_titles = []
for j in new:
    title_index = j-1
    indexes_titles.append(title_index)
```

```python
titles = []
urls = []

for i in new:
    titles.append(lines[i-1])
    urls.append(lines[i])

descs = []
index_descs = []
    
for j in lines:    
    if lines.index(j) not in new and lines.index(j) not in indexes_titles:
        index_descs.append(lines.index(j))
        descs.append(j)
```

```python
index_descs
```

```python
for i in index_descs[0:4]:
    print(lines[i])
```

```python
#remove the repeated '\n' strings
index_descs = list(set(index_descs))
index_descs
```

```python
for i in index_descs:
    if index_descs.index(i) - index_descs.index(i-1) != 1:
        print(i)
```

```python
df = pd.DataFrame(list(zip(titles,indexes_titles,urls,new)),
               columns =['title','title_index','url','url_index'],)
df.index+=1
df['title'] = df['title'].str.replace('\n','')
df['url'] = df['url'].str.replace('\n','')
df['url'] = (df['url'].str.split(' › '))
df['url'] = df['url'].str[0]
df
```
